```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv("insurance.csv")
```

```python
df["sex"]=df["sex"].map({"male" : 1, "female" : 0})
df["smoker"]=df["smoker"].map({"yes" : 1, "no" : 0})
df=pd.get_dummies(df, columns=["region"], drop_first = True, dtype=int)
```

```python
df["underweight"]=(df["bmi"]<18).astype(int)
df["overweight"]=(df["bmi"]>25).astype(int)
df["normalweight"]=((df["bmi"]>=18) & (df["bmi"]<=25)).astype(int)
df["age_smokers"]=(df["smoker"]*df["age"])
df["bmi_smokers"]=(df["smoker"]*df["bmi"])
df
```

```python
x=df.drop(columns=(["charges"]))
y=df["charges"]
```

```python
x_train, x_test, y_train, y_test=train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
pred_test=model.predict(x_test)
```

```python
r2=r2_score(y_test,pred_test)
print("r-Squared:",r2)
```

```python
pred_train=model.predict(x_train)
```

```python
r2_train=r2_score(y_train,pred_train)
print("r-Squared:",r2_train)
```

```python
a=[0.001, 0.01, 0.1, 1.0, 10, 100]
lassocv_model= LassoCV(
   alphas=a,
   cv=5,
   max_iter=1000,
   random_state=42 
)

lassocv_model.fit(x_train, y_train)
```

```python
lasso_pred_test=model.predict(x_test)
print(f"Best Alpha = {lassocv_model.alpha_}")
lasso_r2=r2_score(y_test,lasso_pred_test)
print("r-Squared:",lasso_r2)
mse=mean_squared_error(y_test,lasso_pred_test)
print("MSE = ",mse)
```
